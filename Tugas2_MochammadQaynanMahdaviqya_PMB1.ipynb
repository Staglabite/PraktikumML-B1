```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
plt.rcParams['figure.figsize'] = (16, 9)
# Men-generate dataset yang terkelompok dalam 4 cluster
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Perbaikan di sini
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
plt.show() # Menambahkan plt.show() untuk menampilkan plot
```

```python
import pandas as pd
from sklearn.datasets import load_svmlight_file
import matplotlib.pyplot as plt

# Load the CSV file using pandas
df = pd.read_csv("/content/Mammographic_Mass.csv")

# Assuming the first column is the target and the rest are features
y_train = df.iloc[:, 0].values
X_train = df.iloc[:, 1:].values

# Now you can use X_train and y_train for your model

# Example usage with scatter plot:
plt.scatter(X_train[:, 0], X_train[:, 1])
plt.show()
```

y_train = df.iloc[:, 0].values

X_train = df.iloc[:, 1:].values

Baris ini mengambil semua baris (:) dan kolom pertama (0) dari DataFrame df menggunakan iloc dan menyimpannya dalam variabel y_train. .values digunakan untuk mengonversi data menjadi array NumPy. Ini berarti y_train sekarang berisi nilai-nilai variabel target.

plt.scatter(X_train[:, 0], X_train[:, 1])

Baris ini membuat scatter plot menggunakan Matplotlib.
X_train[:, 0] merujuk pada semua baris (:) dan kolom pertama (0) dari array X_train, yang mewakili nilai-nilai fitur pertama. Ini akan digunakan sebagai koordinat x dari titik-titik dalam plot.
X_train[:, 1] merujuk pada semua baris (:) dan kolom kedua (1) dari array X_train, yang mewakili nilai-nilai fitur kedua. Ini akan digunakan sebagai koordinat y dari titik-titik dalam plot.

```python
X = np.array(list(zip(X_train[:, 0], X_train[:, 1]))).reshape(len(X_train[:, 0]), 2) # Change features to X_train
SSE = []
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    SSE.append(kmeanModel.inertia_)
```

Menjalankan algoritma K-Means untuk nilai k yang berbeda (dari 1 hingga 9) dan menyimpan nilai SSE untuk setiap k. Nilai-nilai SSE ini kemudian dapat digunakan untuk menentukan jumlah cluster yang optimal menggunakan metode elbow. Plot elbow dibuat dengan memplot nilai SSE terhadap nilai k, dan jumlah cluster yang optimal biasanya ditemukan di "elbow" plot, yaitu titik di mana SSE mulai menurun lebih lambat.

```python
plt.plot(K, SSE, 'bx-')
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```

Membuat plot elbow method, yang digunakan untuk menentukan jumlah cluster yang optimal dalam algoritma K-Means. Plot elbow menunjukkan hubungan antara jumlah cluster (k) dan Sum of Squared Errors (SSE). Jumlah cluster yang optimal biasanya ditemukan di "elbow" plot, yaitu titik di mana SSE mulai menurun lebih lambat.

```python
y_pred = KMeans(n_clusters=3).fit_predict(X)
plt.plot
LABEL_COLOR_MAP = {0 : 'r', 1 : 'g', 2 : 'b'}
label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(X[:, 0], X[:, 1], c=label_color)
plt.title("K=3")
plt.show()
```

Melakukan clustering menggunakan algoritma K-Means dengan 3 cluster dan kemudian memvisualisasikan hasil clustering dalam scatter plot di mana setiap cluster diwakili oleh warna yang berbeda.

```python
figure,ax=plt.subplots(2,3)
K=range(1,7)
for k in K:
  if(k<4):
    row=0
    column=k-1
  else:
    row=1
    column=k-4
  kmeanModel=KMeans(n_clusters=k).fit(X)
  y_pred = kmeanModel.fit_predict(X)
  ax[row][column].scatter(X[:, 0], X[:, 1], c=y_pred, cmap='winter_r')

plt.show()
```

Menjalankan algoritma K-Means untuk nilai k yang berbeda (dari 1 hingga 6) dan membuat scatter plot untuk setiap k pada subplot yang terpisah. Hal ini memungkinkan Anda untuk memvisualisasikan hasil clustering untuk beberapa nilai k secara bersamaan dan membantu dalam menentukan jumlah cluster yang optimal menggunakan metode elbow.
